```python
#Import dependencies
import pandas as pd
```

```python
#Import csv file
df = pd.read_csv("Resources/listings.csv")
master_df = pd.DataFrame(df)
master_df["id"].count()
```

```python
#See how many reviews we lose when we look at only those greater than 5 review-we felt this was too many to lose
master_df.loc[df['number_of_reviews']>5]["id"].count()
```

```python
#We decided to use reviews greater than 1 because it got rid of properties with no reviews,
#but it gave us a large enough dataset for analysis
master_df.loc[df['number_of_reviews']>1]["id"].count()
```

```python
#We noticed there were more listings than counts of reviews so we dropped null reviews 
master_df = master_df.loc[df['number_of_reviews']>1]
master_df = master_df[master_df['review_scores_rating'].notnull()]
master_df["id"].count()
```

```python
master_df.describe()
#All of this description is based on properties with more than one listing
#Average review score rating is 96.7
#There are 4081 total listings available in Denver
#There are an average of 2.76 reviews per month
```
